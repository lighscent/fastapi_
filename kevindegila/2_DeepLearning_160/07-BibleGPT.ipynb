# Generer du Texte

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

def data_generator(sequences, vocab_size, max_len, batch_size):
  """
  Generates batches of padded input sequences and one-hot encoded target sequences.

  Args:
    sequences: A list of integer sequences.
    vocab_size: The size of the vocabulary.
    max_len: The maximum length of the input sequences after padding.
    batch_size: The number of samples per batch.

  Yields:
    A tuple containing a batch of padded input sequences and a batch of one-hot
    encoded target sequences.
  """
  while True:
    batch_x = []
    batch_y = []
    for seq in sequences:
      for i in range(1, len(seq)):
        input_seq = seq[:i+1]
        padded_input = pad_sequences([input_seq[:-1]], padding="pre", maxlen=max_len - 1)[0]
        output_token = input_seq[-1]
        one_hot_output = to_categorical([output_token], num_classes=vocab_size)[0]

        batch_x.append(padded_input)
        batch_y.append(one_hot_output)

        if len(batch_x) == batch_size:
          yield np.array(batch_x), np.array(batch_y)
          batch_x = []
          batch_y = []

    if batch_x: # Yield any remaining samples in the last batch
        yield np.array(batch_x), np.array(batch_y)

# Define parameters
vocab_size = len(tokenizer.word_index) + 1 # Add 1 for padding
max_len = 65
batch_size = 2048

# Create the data generator
train_generator = data_generator(sequences, vocab_size, max_len, batch_size)

# Modify the model fitting to use the generator
# Calculate steps per epoch
total_samples = sum([len(seq) - 1 for seq in sequences])
steps_per_epoch = total_samples // batch_size

model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Embedding(vocab_size, 16),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
        tf.keras.layers.Dense(vocab_size, activation='softmax'),
    ]
)

model.compile(loss='categorical_crossentropy', optimizer="adam", metrics=['accuracy'])

h = model.fit(train_generator, epochs=20, steps_per_epoch=steps_per_epoch)

model.summary()
```

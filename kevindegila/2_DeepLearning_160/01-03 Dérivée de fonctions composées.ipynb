```python
def double(x):
  return 2 * x

def carre(x):
  return x**2
```

```python
def compose(x):
  temp = double(x)
  r = carre(temp)
  return r
```

```python
compose(3)
```

```python
def derivee(fonction, point):
  variation = fonction(point+0.0001) - fonction(point)
  taux = variation / 0.0001
  return taux
```

```python
# derivéee en 3 de la composée
```

```python
derivee(double, 3), derivee(carre, 6), derivee(double, 3) * derivee(carre, 6)
```

```python
derivee(compose, 3)
```

```python
def triple_carre(x):
  return 3 * x**2
```

```python
#[double, carre, triple_carre], en 3
```

```python
temp1 = double(3)
temp2 = carre(temp1)
```

```python
derivee(triple_carre, temp2) * derivee(carre, temp1) * derivee(double, 3)
```

```python
def compose(x):
  temp1 = double(x)
  temp2 = carre(temp1)
  return triple_carre(temp2)
```

```python
derivee(compose, 3)
```

# Règle de la Chaine

## g o f'(x) = f'(x) * g'(f(x))

## h o g o f'(x) = f'(x) * g'(f(x)) * h'(g(f(x)))

```python
def chain_rule(chaine, point):

  f = chaine[0]
  g = chaine[1]
  h = chaine[2]


  # forward pass
  f_point = f(point)
  g_point = g(f_point)


  # backward pass
  df_dx = derivee(f, point)
  dg_dx = derivee(g, f_point)
  dh_dx = derivee(h, g_point)


  return df_dx * dg_dx * dh_dx
```

```python
chain_rule([double, carre, triple_carre], 3)
```

```python
def t(x):
  return x**2 - 8
```

```python
chain_rule([double, t, triple_carre], 3)
```

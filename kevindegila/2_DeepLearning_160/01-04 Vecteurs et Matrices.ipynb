```python
maison = np.array([[  2,  40],
       [  4,  35],
       [  3,  21],
       [  3,  18]])
weights = np.array([0.23, 3])
```

```python
maison[0] * weights, maison[0].dot(weights), sum(maison[0] * weights)  # Produit matriciel
```

```python
prix = 0
for x, w in zip(maison, weights):
  prix = prix + x * w
prix
```

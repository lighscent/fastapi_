```python
def f(x):
  return x**2
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x_values = np.linspace(-3, 3, 100)

# Calculate y values using the defined function f(x)
y_values = f(x_values)

# Create the plot
plt.plot(x_values, y_values)

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = x^2')

# Show the plot
plt.grid(True)
plt.show()
```

```python
def derivee(fonction, point):
  variation = fonction(point+0.000001) - fonction(point)
  taux = variation / 0.000001
  return taux
```

```python
derivee(f, 1)
```

# 6/30 - List Comprehension
Les listes comprehension sont une facon habile de créer des listes sur une seule ligne de code, ce qui rend le code beaucoup plus rapide (car python est un langage assez lent)

Les deux code ci-dessous effectuent chacun la meme opération. On peut voir (grace a la commande %%time) que le temps d'execution avec liste comprehenion est bien inférieur au temps d'execution avec la méthode append() 

```python
%%time
liste = []
for i in range(100_000):
  liste.append(i**2)
```

```python
%%time
liste = [i**2 for i in range(100_000)]
```

## Nested list comprehension

```python
liste_1=[i for i in range(3)]
liste_1
```

```python
liste_2 = [[i+j for i in range(3)] for j in range(4)]
liste_2
```

On peut rajouter des conditions **if** dans les listes comprehension, par exemple :

```python
%%time
liste = [i**2 for i in range(100_000) if (i % 2) == 0] # calcule i**2 seulement pour les nombres pairs.

print(liste[-10:]) # affiche les 10 derniers éléments de la liste
```

## Dict comprehension

```python
%%time
prenoms = ['Pier', 'Pol', 'Jack']
users = {k:v for k, v in enumerate(prenoms)}
users
```

```python
users.keys()
```

```python
users.values()
```

```python
users.items()
```

```python
ages = [15,18,21]
users2 = {prenom:age for prenom, age in zip(prenoms, ages)}
users2
```

```python
users3 = {prenom: age for prenom, age in zip(prenoms, ages) if age >= 18}
users3
```

## Tuple comprehension

```python
tup1 = tuple(i**2 for i in range(11))
tup1
# list(enumerate(tup1))
```

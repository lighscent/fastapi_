# 13/30 Numpy et Broadcasting

```python
import numpy as np
import random
```

```python
np.random.seed(0)
A = np.random.randint(0,10, [2, 3])
print(A)
B = np.random.randint(0,10, [2, 1])
print(B) # B sera étendu = Broadcasting
print('-'*77)
C = A + B
print(C)
```

```python
A = np.ones((2, 3))
B = 3
print(A+B) # résultat 1
 
A = np.ones((2, 3))
B = np.ones((2, 1)) # B a une colonne, elle sera étendu sur les trois colonnes de A

print(A+B) # résultat 2
 
# A = np.ones((2, 3))
# B = np.ones((2, 2))
```
